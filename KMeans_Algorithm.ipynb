```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris_data=load_iris()
data=iris_data.data
traget=iris_data.target
df=pd.DataFrame(data,columns=iris_data.feature_names)
df.head(10)
```

```python
squared_error=[]
k_values=range(1,10)

for k in k_values:
    kmeans=KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data)
    squared_error.append(kmeans.inertia_)
plt.plot(k_values,squared_error , marker='o') 
plt.xlabel("Number of clusters (K)")
plt.ylabel("Sum of Squared Errors  (SSE)")
plt.title("Elbow Method to Determine Optimal K")
plt.show()
```

```python
Kmeans=KMeans(n_clusters=3, random_state=42)
predicate_clusters=Kmeans.fit_predict(data)
df['Cluster']=predicate_clusters
print(df.head(7))
```

```python
sns.scatterplot(x=data[:,0],y=data[:,1], hue=predicate_clusters, palette="viridis")
plt.xlabel(iris_data.feature_names[0])
plt.ylabel(iris_data.feature_names[1])
plt.title("KMeans Clustering on Iris DataSet (K=3)")
plt.show()
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix

print("Confusion Matrix :\n",confusion_matrix(traget,predicate_clusters ))
```
